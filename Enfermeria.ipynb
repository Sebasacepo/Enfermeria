```python
#Sebastian Acevedo Restrepo 22258530
import pandas as pd

class SeguroMedico:
    def __init__(self, datos):
        self.datos = datos

    def cargar_datos(self):        
        archivoDF = pd.read_csv(self.datos)
        return(archivoDF)

    def informacion_basica(self):        
        datos = pd.DataFrame(self.cargar_datos())
        filas = datos.shape[0]
        cols = datos.shape[1]
        total = filas * cols #cuenta las filas (0) y columnas(1)
        tipoDato = datos.dtypes
        print(f'Numero de filas: {filas} \nNumero de columnas: {cols}\ntotal de datos: {total}\n')
        print(f'tipos de dato": {tipoDato}')     

    def información_general(self):
        datos=self.cargar_datos()
        filas = datos.shape[0]
        edadProme = (datos['age'].sum())/ filas
        hijosProme = (datos['children'].sum() /filas)
        
        print(f'Edad promedio: {edadProme}\nHijos promedio: {hijosProme}')

    def filtrar_personas(self,edad, hijos):
        tabla= self.cargar_datos()
        colHijos = tabla.loc[:, [hijos, edad]]
        print(f'Cantidad de personas: {colHijos.shape[0]}')      
        return (colHijos)
    
    def pacientes_que_fuman(self,consulta, consulta2):       
        tabla = pd.DataFrame(self.cargar_datos())
        noFuman = tabla.query(consulta2) 
        siFuman = tabla.query(consulta)
        print(f'Cantidad de no fumadores: {noFuman.shape[0]}\nCantidad de fumadores: {siFuman.shape[0]}')
        return noFuman,siFuman
    
    def analisis_costos(self, siFumador, noFumador):
        promSiFum = siFumador['charges'].sum() / siFumador.shape[0]
        promNoFum = noFumador['charges'].sum() / noFumador.shape[0]
        print(f'Costo promedio de los fumadores: {promSiFum}\nCosto promedio de los no fumadores: {promNoFum}')

    def costos_por_region():
        pass

    def filtrar_por_condicion():
        pass

    def  filtrado_especial(self, siFuman, consulta):
        #El filtro sera mujeres que tengas hijos, mayores de 30 y que vivan en southeast
        tabla = self.cargar_datos()
        consulta = tabla.query(consulta)
        return consulta

    def costos_por_region(self):
        tabla = self.cargar_datos()
        promeRegion = tabla.groupby('region')['charges'].mean()
        maxRegio = promeRegion.idxmax()        
        print(f"La región donde más cuesta pagar un seguro médico es '{maxRegio}'\n Promedio de por {promeRegion}")
    
    def filtrar_por_condicion(self, sexo, region):
        tabla = self.cargar_datos()
        return tabla.query(f'sex == "{sexo}" and region == "{region}"')

    def exportar_a_csv(self):       
        pass

    def ingresar_paciente(self, datos_paciente):       
        pass

        

        
            
```
